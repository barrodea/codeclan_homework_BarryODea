```python
# Q3 
drinks = drinks_content[["Beverage_category", "Beverage", "Beverage_prep", "Calories"]].copy()
drinks.dropna()
```

```python
# Q4 
drinks["calorie_diff"] = drinks.Calories - 135
drinks
```

```python
# Q5 
drinks.groupby("Beverage_category").mean()
```

```python
# Q6 
drinks_mean_calories = drinks.Calories.mean()
drinks_mean_calories
```

```python
drinks.groupby("Beverage_prep").mean()
```

```python
# Q7 
pd.unique(drinks.Beverage_category)

coffee_drinks = drinks.loc[drinks["Beverage_category"].isin(["Coffee", "Classic Espresso Drinks", "Signature Espresso Drinks",
       "Frappuccino® Blended Coffee", "Frappuccino® Light Blended Coffee"])].copy()


coffee_drinks.groupby("Beverage_category")["Calories"].min()
```

```python
# Q8 
espresso_drinks = drinks.loc[drinks["Beverage_category"] == "Classic Espresso Drinks"].copy()
espresso_drinks.head(5)
```

```python
# Q9 
espresso_drinks.groupby("Beverage_prep")["Calories"].mean()
```

```python
# Q10
espresso_drinks.groupby("Beverage_prep")["Calories"].mean().reset_index(name = "mean_claories")
```
